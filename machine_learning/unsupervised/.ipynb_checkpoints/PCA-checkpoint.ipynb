```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

# Principal Component Analysis (PCA)

* rotates the dataset in a way such that the rotated features are statistically uncorrelated.
* followed by selecting only a subset of the new features, according to how important they are for explaining the data.

PCA on 2D synthetic data:

![title](img/pca.png)

* The algorithm proceeds by **first** finding the **direction of maximum variance**, labeled “Component 1.” This is the direction (or vector) in the data that contains most of the information, or in other words, the direction along which the features are most correlated with each other.
* Then, the algorithm finds the direction that contains the most information while being orthogonal (at a right angle) to the first direction.
* In 2D (plane) there is only one perpendicular degree of freedom but in N dimensions there are (N-1) possibile orthogonal directions.
* The direction of the arrows are not important, only the alignment.
* The directions found using this process are called **principal components**, as they are the main directions of variance in the data.
* In general, there are as many principal components as original features.
* The **mean** is subtracted from each data point and then rotated - centered around 0.
* The second plot is rotated so that the "Component 1" is aligned with the x-axis and "Component 2" with the y-axis
* In the rotated representation found by PCA, the two axes are uncorrelated, meaning that the correlation matrix of the data in this representation is zero except for the diagonal.
* We can **use PCA for dimensionality reduction by retaining only some of the principal components**.

![title](img/pca2.png)

* We extracted only the "Component 1", since we found it most ineteresting.
* Added back the means and rotated back to original angle, resulting in figure to the right.
* These points are in the original feature space, but we kept only the information contained in the first principal component

## First Principial Component
 * plot two features on scatter plot
 * find correlation coefficent
 * plot the alignment/direction of the first principial component of these features

```python
grain = pd.read_csv('data/seeds.csv', header=None, 
                    usecols=[3, 4, 7], names=['length', 'width', 'variety'])

names = ['Kama wheat', 'Rosa wheat', 'Canadian wheat']
grain['var_names'] = grain['variety'].replace({
    1: names[0],
    2: names[1],
    3: names[2]
})
grain.head(2)
```

```python
# the data
X = grain[['length', 'width']].values

# pearson correlation coefficient
r = pearsonr(X[:,0], X[:,1])
print("Pearson R = {:.3f}".format(r[0]))

# finding the first principial component. Its the one, where the
# data varies the most.
pca = PCA(n_components=1)
pca.fit(X)

# length and width means of the grain samples
means = pca.mean_

# get the 1st principial component: first_pc
first_pc = pca.components_[0,:]

# plotting grain lengths and widths
plt.figure(figsize=(9, 5))
plt.scatter(x=grain.length.values, y=grain.width.values, 
            c=grain.variety, cmap='viridis')
plt.axis('equal')
plt.xlabel('Grain Length')
plt.ylabel('Grain width')
plt.title('Grain length vs grain width', weight='bold')

# Plot first_pc as an arrow, starting at mean
plt.arrow(means[0], means[1], first_pc[0], first_pc[1], color='red', width=0.03)

plt.show()
```

## Intrinsic dimension
- Intrinsic dimension = number of features needed to approximate the dataset
- Essential idea behind dimension reduction
- What is the most compact representation of the samples?
- Can be detected with PCA
- Intrinsic dimension is dimension representation of the data which is usually reduced and not much information is lost. 
- This is done to reduce the dimensions of the dataset.

#### Determine Fish Dataset Intrinsic dimension

```python
# import data
fish = pd.read_csv('data/fish.csv', header=None, 
                   names=['target', 'a', 'b', 'c', 'd', 'e', 'f'])
fish.head(1)
```

```python
X = fish.drop('target', axis=1).values

# scaler, estimator and pipeline obj
scaler = StandardScaler()
pca = PCA()
steps = [('scaler', scaler), ('pca', pca)]

pipe = Pipeline(steps=steps)
# fit samples
pipe.fit(X)

# Plot the explained variances
features = range(pca.n_components_)
plt.figure(figsize=[6,4])
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.title('Variances of principial components in Grain dataset', weight='bold')
plt.show()
```

It looks like PCA features 0 and 1 have significant variance. Since PCA features 0 and 1 have significant variance, the intrinsic dimension of this dataset appears to be **2**. PCA features are in decreasing order of variance.

## Dimension reduction with PCA

- Represents same data, using less features
- Important part of machine-learning pipelines
- Can be performed using PCA
- PCA features are in decreasing order of variance
- Assumes the low variance features are "noise"
- ... and high variance features are informative
- Specify how many features to keep
- E.g. PCA(n_components=2)
- Keeps the first 2 PCA features
- Intrinsic dimension is a good choice

#### Reducing Iris dataset dimensions

```python
iris = datasets.load_iris()
X = iris.data

scaler = StandardScaler()
pca = PCA()
steps = [('scaler', scaler), ('pca', pca)]

pipe = Pipeline(steps)
pipe.fit(X)

# Plot the explained variances
features = range(pca.n_components_)
plt.figure(figsize=[6,4])
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.title('Variances of principial components in Iris dataset', weight='bold')
plt.show()
```

The intrinsic dimension seems to be 2. Let's reduce it to 2.

```python
pca = PCA(n_components=2).fit(X)

X_transformed = pca.transform(X)
print('The shape of original dataset: {}'.format(X.shape))
print('The shape of the transformed dataset: {}'.format(X_transformed.shape))
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.figure(figsize=(8, 6))
plt.scatter(X_transformed[:,0], X_transformed[:,1], 
            c=iris.target, cmap='viridis')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('Iris data as PCs')
plt.show()
```

#### Reducing Breast Cancer dataset dimensions

```python
cancer = datasets.load_breast_cancer()

# init scaler obj, fit & transform
scaler = StandardScaler()
scaler.fit(X=cancer.data)
X_scaled = scaler.transform(cancer.data) # each feature has unit variance

# init pca obj & keep first 2 principal components of the data
pca = PCA(n_components=2)
pca.fit(X_scaled)

# apply rotation and dim-reduction
X_pca = pca.transform(X_scaled)

print('Original shape: {}'.format(X_scaled.shape))
print('Transformed shape: {}'.format(X_pca.shape))
```

```python
cancer.target_names
```

```python
# plotting the first 2 principal components colored by class

_ = plt.figure(figsize=(8,6))
_ = plt.scatter(X_pca[:,0], X_pca[:,1], c=cancer.target, edgecolor='black', s=50)
_ = plt.xlabel('1st Principal Component')
_ = plt.ylabel('2nd Principal Component')
_ = plt.title('First 2 Principal Components of Cancer Dataset', weight='bold')
plt.show()
```

* the two classes separate quite well in this two-dimensional space
* This leads us to believe that even a linear classifier (that would learn a line in this space) could do a reasonably good job at distinguishing the two classes
* maligant (blue) points are more spread out
* A downside of PCA is that the two axes in the plot are often not very easy to interpret.
* The principal components correspond to directions in the original data, so they are combinations of the original features.
* The principal components coefficients themselves are stored in the **components_** attribute of the PCA object during fitting:

```python
print("PCA components:\n{}".format(pca.components_))
```

* Each row in components_ corresponds to one principal component, and they are sorted by their importance (the first principal component comes first, etc.). 
* The columns correspond to the original features attribute of the PCA in this example, “mean radius"

We can visualize the componetns with the ehatmap:

```python
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ["First component", "Second component"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)),
    cancer.feature_names, rotation=60, ha='left')
plt.xlabel("Feature")
plt.ylabel("Principal components")
plt.show()
```

* In the first component all features have the sam sign - general correlation. If one feature goes up the others do aswell.
* The second component has mixed signs, which makes explaining the axes in previous figure complicated.

## Face Recognition with PCA

* Dataset of celebrity images
* We use gray‐scale versions of these images, and scale them down for faster processing

```python
people = datasets.fetch_lfw_people(min_faces_per_person=20, resize=0.7)

image_shape = people.images[0].shape

fix, axes = plt.subplots(2, 5, figsize=(15, 8),
                         subplot_kw={'xticks': (), 'yticks': ()})
plt.set_cmap('Greys_r')
for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
plt.show()
```

```python
print("people.images.shape: {}".format(people.images.shape))
print("Number of classes: {}".format(len(people.target_names)))
```

* There are 1682 images.
* Each image 87x65 pixels
* 20 differnet persons

```python
# counts of imeages per person
counts = np.bincount(people.target)

# print counts next to target names
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print("{0:25} {1:3}".format(name, count), end='  ')
    if (i + 1) % 3 == 0:
        print()
```

* The dataset is a bit skewed, however, containing a lot of images of George W. Bush and Colin Powell
* we will only take up to 50 images of each person

```python
mask = np.zeros(people.target.shape, dtype=np.bool)

for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
    
X_people = people.data[mask]
y_people = people.target[mask]

# scale the grayscale values to be between 0 and 1
# instead of 0 and 255 for better numeric stability
X_people = X_people / 255.
```

* A common task: if a previously unseen face belongs to a known person from a database. 
* One way to solve this problem would be to build a classifier where each person is a separate class. However, there are usually many different people in face databases, and very few images of the same person. That makes it hard to train most classifiers. One also wants to add people without retrain a large model.
* A simple solution is to use a one-nearest-neighbor classifier that looks for the most similar face image to the face you are classifying.
* This classifier could in principle work with only a single training example per class.

Check how well kNN would perform:

```python
X_train, X_test, y_train, y_test = train_test_split(X_people, y_people,
                                                   stratify=y_people,
                                                   random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print("Test set accuracy with 1-nn: {:.3f}".format(knn.score(X_test, y_test)))
```

* Computing distances in the original pixel space is quite a bad way to measure similarity between faces.
* We hope that using distances along principal components can improve our accuracy
* we enable the **whitening** option of PCA, which rescales the principal components to have the same scale. This is the same as using StandardScaler after the transformation.
* **whitening** corresponds to not only rotating the data, but also rescaling it so that the center panel is a circle instead of an ellipse:

![title](img/pca3.png)

We fit the PCA object to the training data and extract the first 100 principal components. Then we transform the training and test data:

```python
pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train, y_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("X_train.shape: {}".format(X_train.shape))
print("X_train_pca.shape: {}".format(X_train_pca.shape))
```

* The new data has 100 features, the first 100 principal components. 
* we can use the new representation to classify our images using a one-nearest-neighbors classifier:

```python
knn = KNeighborsClassifier(n_neighbors=1).fit(X_train_pca, y_train)

print("Test set accuracy with PCA tuned: {:.3f}".format(knn.score(X_test_pca, y_test)))
```

* For image data, we can also easily visualize the principal components that are found.
* Components correspond to directions in the input space. 
* The input space here is 50×37-pixel grayscale images, so directions within this space are also 50×37-pixel grayscale images.

Let’s look at the first couple of principal components

```python
print("pca.components_.shape: {}".format(pca.components_.shape))
```

```python
fix, axes = plt.subplots(3, 5, figsize=(15, 12),
                         subplot_kw={'xticks': (), 'yticks': ()})

for i, (component, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape),
              cmap='viridis')
    ax.set_title("{}. component".format((i + 1)))
plt.show()
```

## Word Frequency Arrays [TruncatedSVD]

- Each row represents a document, each column represents a particular word
- Entiries measure presence of each word in each document 
- Most entires are 0 though since many words are not included in any document
- Arrays like these are said to be **'sparse'** - most entries are 0.
- This case scipy.sparse.csr_matrix can be used for an alternative to NumPy array
- csr_matrixes saves space since it remembers non-zero entries only
- scikit-learn PCA doesn't support csr_matrixes
- have to use scikit-learn TruncatedSVD instead 
- TruncatedSVD performs same transformations as PCA but accepts **csr_matrixes** as input

```python
documents = ['cats say meow', 'dogs say woof', 'dogs chase cats']

# Create a TfidfVectorizer: tfidf
# Convert a collection of raw documents to a matrix of TF-IDF features.
tfidf = TfidfVectorizer()

# Apply fit_transform to document: csr_mat
csr_mat = tfidf.fit_transform(documents)

print(documents, '\n')

# Print result of toarray() method
print(np.around(csr_mat.toarray(), decimals=2))

# Get the words: words
words = tfidf.get_feature_names()

# Print words
print('\n{}'.format(words))
```
