# Puzzle 21
## Part I
--- Day 21: Dirac Dice ---
There's not much to do as you slowly descend to the bottom of the ocean. The submarine computer challenges you to a nice game of Dirac Dice.

This game consists of a single die, two pawns, and a game board with a circular track containing ten spaces marked 1 through 10 clockwise. Each player's **starting space** is chosen randomly (your puzzle input). Player 1 goes first.

Players take turns moving. On each player's turn, the player rolls the die **three times** and adds up the results. Then, the player moves their pawn that many times **forward** around the track (that is, moving clockwise on spaces in order of increasing value, wrapping back around to 1 after 10). So, if a player is on space 7 and they roll 2, 2, and 1, they would move forward 5 times, to spaces 8, 9, 10, 1, and finally stopping on 2.

After each player moves, they increase their **score** by the value of the space their pawn stopped on. Players' scores start at 0. So, if the first player starts on space 7 and rolls a total of 5, they would stop on space 2 and add 2 to their score (for a total score of 2). The game immediately ends as a win for any player whose score reaches **at least 1000**.

Since the first game is a practice game, the submarine opens a compartment labeled **deterministic dice** and a 100-sided die falls out. This die always rolls 1 first, then 2, then 3, and so on up to 100, after which it starts over at 1 again. Play using this die.

For example, given these starting positions:
<pre>
Player 1 starting position: 4
Player 2 starting position: 8</pre>
This is how the game would go:
<pre>
Player 1 rolls 1+2+3 and moves to space 10 for a total score of 10.
Player 2 rolls 4+5+6 and moves to space 3 for a total score of 3.
Player 1 rolls 7+8+9 and moves to space 4 for a total score of 14.
Player 2 rolls 10+11+12 and moves to space 6 for a total score of 9.
Player 1 rolls 13+14+15 and moves to space 6 for a total score of 20.
Player 2 rolls 16+17+18 and moves to space 7 for a total score of 16.
Player 1 rolls 19+20+21 and moves to space 6 for a total score of 26.
Player 2 rolls 22+23+24 and moves to space 6 for a total score of 22.</pre>
...after many turns...
<pre>
Player 2 rolls 82+83+84 and moves to space 6 for a total score of 742.
Player 1 rolls 85+86+87 and moves to space 4 for a total score of 990.
Player 2 rolls 88+89+90 and moves to space 3 for a total score of 745.
Player 1 rolls 91+92+93 and moves to space 10 for a final score, 1000.</pre>

Since player 1 has at least 1000 points, player 1 wins and the game ends. At this point, the losing player had 745 points and the die had been rolled a total of 993 times; 745 * 993 = **739785**.

Play a practice game using the deterministic 100-sided die. The moment either player wins, **what do you get if you multiply the score of the losing player by the number of times the die was rolled during the game?**

```python
# initial positions of the dest data
p1_0 = 4
p2_0 = 8

def score(p01, p02):

    # generate die rolls
    ids = [i for i in range(0,301,3)]
    rolls = [i for i in range(1,101)] * 3
    rolls3 = [sum(rolls[i:i+3]) for i in ids]
    
    # init dict: 'player': position, score, n_rolls
    dct = {'p1':[p01,0], 'p2': [p02,0]}
    tot_rolls = 0
    while True:
        try:
            for i,sum3 in enumerate(rolls3):
                p = 'p1' if i % 2 == 0 else 'p2'
                dct[p][0] = 10 if (dct[p][0]+sum3) % 10 == 0 else (dct[p][0]+sum3) % 10
                dct[p][1] += dct[p][0]
                tot_rolls += 3
                if dct[p][1] >= 10:
                    raise StopIteration
        except StopIteration: break
             
    return dct, tot_rolls

score(4,8)
```

```python
def n_cubes_on(data_file, only_init=False):
    """Compute number of cubes witched 'on'."""
    # read in data
    df = pd.read_csv(data_file, header=None,
                     sep=r"\sx=|\.\.|,y=|,z=|,$", engine='python',
                     names=['logic', 'x1', 'x2', 'y1', 'y2', 'z1', 'z2'])
    
    # only initialization or full data
    if only_init == True:
        df = df.iloc[:20] 
    
    # find the min value of the grid
    origin = abs(df.loc[:,'x1':'z2'].min().min())
    
    # add size to all data points to shift origin to positive
    df.loc[:,'x1':'z2'] = df.loc[:,'x1':'z2'] + origin
    
    # find the size of the reactor
    size = df.loc[:,'x1':'z2'].max().max()
    
    # convert 'on' to 1 and 'off' to 0
    df['logic'] = df.logic.apply(lambda x: 1 if x=='on' else 0)
    
    # create 3D numpy arry with all zeros
    reactor = np.zeros((size+1,size+1,size+1), dtype=int)
    
    # switch cubes on/off in the reactor
    for idx in range(df.shape[0]):
        srs = df.iloc[idx,:]
        
        reactor[srs['x1']:srs['x2']+1,
                srs['y1']:srs['y2']+1,
                srs['z1']:srs['z2']+1] = srs['logic']
        
    print(f"{reactor.sum()} cubes are on in the reactor.")

n_cubes_on('data/pz22_test1.txt', only_init=True)
```

```python
def move_herd(np_array, herd='>', n_moves=0):
    """Move herd, return moved layout and number of moves."""

    m, n = np_array.shape[0], np_array.shape[1]
    new_arr = np_array.copy()
    if herd == ">":
        for i in range(m):
            for j in range(n):
                if np_array[i,j] == herd:
                    if j == n - 1:
                        if np_array[i,0] == '.':
                            new_arr[i,j] = '.'
                            new_arr[i,0] = herd
                            n_moves += 1
                    else:
                        if np_array[i,j+1] == '.':
                            new_arr[i,j] = '.'
                            new_arr[i,j+1] = herd
                            n_moves += 1
        return new_arr, n_moves
    
    elif herd == "v":
        for i in range(m):
            for j in range(n):
                if np_array[i,j] == herd:
                    if i == m - 1:
                        if np_array[0,j] == '.':
                            new_arr[i,j] = '.'
                            new_arr[0,j] = herd
                            n_moves += 1
                    else:
                        if np_array[i+1,j] == '.':
                            new_arr[i,j] = '.'
                            new_arr[i+1,j] = herd
                            n_moves += 1
        return new_arr, n_moves
            
def sea_cucumber(data_file):
    """"""
    data = np.genfromtxt(data_file, delimiter=1, dtype=str)
    
    steps = 0
    while True:
        data, n_moves = move_herd(data, herd='>') # east herd 
        data, n_moves = move_herd(data, herd='v', n_moves=n_moves)        
        steps += 1
        
        if n_moves == 0:
            break
            
    print(f"First step when no sea cucumbers move: {steps}")

sea_cucumber("data/pz25_test.txt")
```

```python
def v2(data_file):
    
    data = np.genfromtxt(data_file, dtype=str)
    
    count = 0
    while True:
        init = data.copy()
        for herd in ('>', 'v'):
            if herd == 'v': 
                data = ["".join(i) for i in zip(*data)]
                    
            data = [i+'.' if (i[0]=='.' and i[-1]==f'{herd}') else i+'x' for i in data]
            data = np.char.replace(data,f"{herd}.", f".{herd}")
            data = [i[:-1] if i[-1]=='x' else i[-1]+i[1:-1] for i in data]

            if herd == 'v': 
                data = ["".join(i) for i in zip(*data)]

        count += 1
        if list(init) == data:
            break
            
    print(f"Cucumbers stop moving after {count} steps.")
v2("data/pz25_test.txt")
```
