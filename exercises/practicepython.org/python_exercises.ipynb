```python
def game_board(size):
    '''Returns size x size board'''
    
    dashes = "".join([' ---' for i in range(size)])
    vbars = "".join(['|   ' for i in range(size + 1)])
    
    board = [dashes, vbars]
    
    for x in range(size):
        for i in board:
            print(i)
    print(board[0])
    
game_board(3)
```

```python
def numbered_game_board(size):
    '''Prints sizexsize game board with placeing numbers.'''
    place_holders = [str(i) for i in range(1, size * size + 1)]
    
    dashes = "".join([' ---' for i in range(size)])
    vbars = [' ' + i + ' |' if int(i) <= 9 else i + ' |' for i in place_holders]

    for i in range(0, len(vbars), size):
        if i == 0:
            print(dashes)
            print('|' + "".join(vbars[i:size]))
        else:
            print(dashes)
            print('|' + "".join(vbars[i:i + size]))
    print(dashes)
numbered_game_board(3)
```

```python
from numpy import copy
# random nested list for Tic Tac Toe game and
# convert it into ndarray
# grid shape should be: np.array([[1, -1, 0], [-1, 1, 0], [1, 1, 1]]) 
# aka nested ndarray

def check_grid(grid, size):
    '''Checking the game board if someone has won.'''
    
    # extract main and second diagonals of the matrix
    main_diag = grid.diagonal()
    sec_diag = np.fliplr(grid).diagonal()

    # extract columns and rows of the matrix
    cols = [grid[:,i] for i in range(size)]
    rows = [grid[i, :] for i in range(size)]

    # putting all possible lines together
    all_elem = [main_diag, sec_diag]

    for x, y in zip(cols, rows):
        all_elem.append(x)
        all_elem.append(y)

    # checking the winner
    for i in all_elem:
        if sum(i) == -3:
            print('Player 1 won!')
            return 1
        elif sum(i) == 3:
            print('Player 2 won!')
            return 1
```

```python
import numpy as np

def tic_tac_toe(size):
    '''Game of Tic Tac Toe'''
    
    place_holders = [' '] * (size ** 2)
    grid = np.zeros(size ** 2, dtype=int)
    players = ['Player X: ', 'Player O: ']
    
    n = 0
    end = False
    
    numbered_game_board(size)
    
    while end == False:
        
        for i,j in zip(range(2), players):
            
            a = int(input(players[i]))
            
            if i == 0:
                place_holders[a-1] = "X"
                grid[a-1] = -1
            elif i == 1:
                place_holders[a-1] = "O"
                grid[a-1] = 1
            
            dashes = "".join([' ---' for i in range(size)])
            vbars = [' ' + place_holders[i] + ' |' for i in range(size ** 2)]

            for i in range(0, len(vbars), 3):
                if i == 0:
                    print(dashes)
                    print('|' + "".join(vbars[i:size]))
                else:
                    print(dashes)
                    print('|' + "".join(vbars[i:i + size]))
            print(dashes)
            
            n += 1
            
            # checking for a winner
            matrix = np.reshape(grid, newshape=(3, 3))
            result = check_grid(grid=matrix, size=size)
            
            if result == 1:
                end = True
                break
        if n == 8:
            break

tic_tac_toe(3)
```

This exercise is Part 1 of 4 of the birthday data exercise series. The other exercises are: Part 2, Part 3, and Part 4.

For this exercise, we will keep track of when our friend’s birthdays are, and be able to find that information based on their name. Create a dictionary (in your file) of names and birthdays. When you run your program it should ask the user to enter a name, and return the birthday of that person back to them.

```python
import pandas as pd

'''Converting web html table into DF'''

# convert a html table into DF
df = pd.read_html(io='https://en.wikipedia.org/wiki/Wikipedia:Database_reports/Birthday_today',
                 header=0)[0]

bdays = df.loc[:, 'Born': 'description']

inds_to_del = []

# searching for datetimes that are invalid
for index, elem in enumerate(bdays['Born']):
    if len(str(elem)) > 10:
        inds_to_del.append(index)
    else:
        continue

# dropping invalid birth dates
clean_bdays = bdays.drop(inds_to_del)
clean_bdays.set_index('Person', inplace=True)
clean_bdays.drop(labels='description', axis='columns', inplace=True)

# convert DF to dictionary
bdays_d = clean_bdays.to_dict()
bdays = bdays_d['Born']
```
