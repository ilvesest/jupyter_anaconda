# Recursion

Find all paths in graph from **<code>start</code>** to **<code>end</code>** only visiting each node once.

```python
G2 = G.copy()
G2.add_edges_from(edges)

# draw graph
plt.figure(figsize=(3,3))
nx.draw(G2, with_labels=True)
plt.show(fig)
```

```python
def traverse_graph(graph):
    paths=[]
    # traverse graph
    def traverse_graph_active(graph, path=['start']):
        """Recursive loop through graph object for
        finding unique paths from 'start' to 'end'.

        Parameters
        ----------
        graph : networkx graph object
        path : list of already visited nodes

        Returns
        -------
        paths : list of all possible paths
        """
        for node in graph.neighbors(path[-1]):
            if not node in path:
                if node == 'end':
                    paths.append(path + ['end'])
                else: 
                    traverse_graph_active(graph, path + [node])
        return paths
    
    
    paths_copy = traverse_graph_active(graph).copy()
    paths.clear # clear paths variable
    return paths_copy

def solve_grapxh(graph, print_paths=False):
    paths = traverse_graph(graph)
    
    print(f"Number of possible paths: {len(paths)}")
    
    if print_paths:
        print("")
        for path in paths:
            print(path)
            
solve_graph(G2, True)
```

# Minimal Cost Path
Find "best" path between 2 nodes in a graph, where "best" represents certain condition. Example: Find path from top left to bottom right with minimal risk level (without counting the risk level of the starting node (0,0)): 

```python
print(aoc_pz15)
```

```python
def graph_min_risk(data, draw=True, figsize=(12,12)):
    """Find best path with networkX Dijstra's algorithm."""
    
    # data shape m x n
    m, n = data.shape[0], data.shape[1]
    
    ob = [-1, m]
    D = nx.DiGraph() # init directed graph
    for i,j in np.ndindex(data.shape):
        
        all_neighbors = [(i,j+1), (i+1,j), (i,j-1), (i-1,j)]
        valid_neighbors = [i for i in all_neighbors if not any(x in i for x in ob)]
        
        # add edges with weights to graph
        for neighbor in valid_neighbors:
            D.add_edge((i,j), neighbor, weight=data[i,j])

    # find shortest path list of nodes
    shortest_path = nx.dijkstra_path(D, source=(0,0), target=(m-1, n-1))
    lowest_risk = sum([data[node] for node in shortest_path if node != (0,0)])
    
    if draw:
        #Node values are currently their coordinates in a grid,
        #replace them with theior face values
        for node in D.nodes():
            D.nodes[node]['val'] = data[node]
            D.nodes[node]['pos'] = node

        # highlight shortest path with green color
        color_map = ['green' if node in shortest_path else 'skyblue' for node in D]
        
        # generate plot
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
        fig.suptitle(f"({m} x {n}) Grid Shortest Weighted Path\nTotal risk: {lowest_risk}", 
                  c='Black', weight='bold', fontsize=16)
        
        pos = nx.get_node_attributes(D, 'pos')
        labels = nx.get_node_attributes(D, 'val')
        edge_labels = nx.get_edge_attributes(D, 'weight')
        
        nx.draw(D, pos, labels=labels, node_color='skyblue', ax=ax1) 
        nx.draw(D, pos, labels=nx.get_node_attributes(D, 'val'), 
                node_color=color_map, ax=ax2)
        
        fig.text(.5, .05, "Grid is flipped but path and numbers are correct.", 
                 ha='center', weight='bold',fontsize=16)
        plt.show()
    else:
        print(f"({m} x {n}) Grid Shortest Weighted Path\nTotal risk: {lowest_risk}")

graph_min_risk(aoc_pz15, figsize=(12,6))
```
